```python
#BUBBLE SORT
def bubble_sort(arr):
    n = len(arr)
    # Melakukan iterasi untuk setiap elemen
    for i in range(n):
        # Elemen terakhir i sudah berada di posisi yang benar
        for j in range(0, n - i - 1):
            # Tukar jika elemen saat ini lebih besar dari elemen berikutnya
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Contoh penggunaan
data = [64, 34, 25, 12, 22, 11, 90]
# Menggunakan .copy() agar data asli tidak berubah
sorted_data = bubble_sort(data.copy())

print("Data sebelum sorting:", data)
print("Data setelah Bubble Sort:", sorted_data)
```

```python
#SELECTION SORT
def selection_sort(arr):
    n = len(arr)
    # Iterate through all array elements
    for i in range(n - 1):
        # Find the minimum element in the remaining unsorted portion
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j

        # Swap the found minimum element with the first element of the unsorted part
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

# Example usage
data = [64, 25, 12, 22, 11]
sorted_data = selection_sort(data.copy())

print("Data sebelum sorting:", data)
print("Data setelah Selection Sort:", sorted_data)
```

```python
#INSERTION SORT
def insertion_sort(arr):
    n = len(arr)  # Fixed: Added assignment '='
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        # Fixed: Condition j >= 0 to include index 0
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1  # Fixed: Corrected decrement from 'j = 1' to 'j -= 1'
        arr[j + 1] = key  # Fixed: Syntax error and correct placement
    return arr

# Example usage
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = insertion_sort(data.copy()) # Fixed: Added assignment '='
print("Data sebelum sorting:", data)
print("Data setelah Insertion Sort:", sorted_data)
```

```python
#MERGE SORT
def merge_sort(arr):
  if len(arr) > 1:
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    merge_sort(left_half)
    merge_sort(right_half)

    i = j = k = 0

    while i < len(left_half) and j < len(right_half):
      if left_half[i] < right_half[j] :
        arr[k] = left_half[i]
        i += 1
      else:
        arr[k] = right_half[j]
        j += 1
        k += 1

    while i < len(left_half):
      arr[k] = left_half[i]
      i += 1
      k += 1

    while j < len(right_half):
      arr[k] = right_half[j]
      j += 1
      k += 1

  return arr

#contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = merge_sort(data.copy())
print("data sebelum sorting:", data)
print("datas setelah merge sort:", sorted_data)

```

```python
#QUICK SORT
def quick_sort(arr):
  if len(arr) <= 1:
    return arr

  pivot = arr[len(arr) // 2]
  left = [x for x in arr if x < pivot]
  middle = [x for x in arr if x == pivot]
  right = [x for x in arr if x > pivot]

  return quick_sort(left) + middle + quick_sort(right)

#contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = quick_sort(data.copy())
print("data sebelum sorting:", data)
print("data setelah quick sort:", sorted_data)
```
